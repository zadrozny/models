# Most common two thousand words and their phonetic transcriptions

```python
top_2000 = {}
for word in all_words:
    try:
        if eval(word[-1])['GUT']['RANK'] < 2000:
            top_2000[word[1]] = eval(word[2])['WIKTIONARY']['Pronunciation'][0][0]
    except:
        pass
```

```python
print(set(list(''.join(top_2000.values()))))
```

```python
for phoneme in sorted(Counter(''.join(top_2000.values())), key=len):
    print(phoneme)
    for word, trans in top_2000.items():
        if phoneme in trans:
            print('    ', '%-30s' % trans, '%-40s' % word)
```
